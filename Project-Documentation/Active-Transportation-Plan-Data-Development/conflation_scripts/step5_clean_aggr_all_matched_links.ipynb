```python
import os, glob
import fiona
import geopandas as gpd
import pandas as pd
```

```python
box_folder = 'C:\\Users\\{}\\Box\\DataViz Projects\\Spatial Analysis and Mapping\\Active Transportation Plan\\Data'.format(os.getenv('USERNAME'))
geojson_folder = os.path.join(box_folder, 'geojson')
shst_match_folder = os.path.join(box_folder, 'shst_match')
shst_clean_folder = os.path.join(box_folder, 'shst_match_results', 'by_data_source_matched')
```

```python
def read_shst_extract(path, suffix):
    """
    Read all matched records regardless of the matching rules.
    """
    
    shst_gdf = pd.DataFrame()
    
    shst_file = glob.glob(path + '/**/' + suffix, recursive = True)
#     print(shst_file)
    print('----------start reading shst extraction data-------------')
    for i in shst_file:
        print('reading shst extraction data : ', i)
        new = gpd.read_file(i)
        new['source'] = i.split('shst_match\\')[1]
        shst_gdf = pd.concat([shst_gdf, new],
                             ignore_index = True,
                             sort = False)
    print('----------finished reading shst extraction data-------------')
    
    shst_gdf.rename(columns = {'shstFromIntersectionId': 'fromIntersectionId',
                               'shstToIntersectionId'  : 'toIntersectionId'},
                    inplace = True)
    return shst_gdf
```

### oakland

```python
suffix = 'oakland_bike_network_epsg4326.out.matched.geojson'

oakland_match_gdf = read_shst_extract(shst_match_folder, suffix)

print(oakland_match_gdf.shape)
print(list(oakland_match_gdf))
display(oakland_match_gdf.head(3))

# check potential duplicates
# oakland_match_gdf_chk = oakland_match_gdf.copy()
# oakland_match_gdf_chk['test'] = oakland_match_gdf_chk['shstReferenceId'] + '_' + oakland_match_gdf_chk['shstGeometryId'] + '_' + oakland_match_gdf_chk['fromIntersectionId'] + '_' + oakland_match_gdf_chk['toIntersectionId'] + '_' + oakland_match_gdf_chk['pp_mtc_facility_id'].astype(str)
# oakland_match_gdf_chk['cnt'] = oakland_match_gdf_chk.groupby(['test'])['test'].transform('count')
# chk = oakland_match_gdf_chk.loc[oakland_match_gdf_chk.cnt>1]
# chk

unique_oakland_match_gdf = oakland_match_gdf.drop_duplicates(
    subset = ['shstReferenceId', 'shstGeometryId', 'fromIntersectionId', 'toIntersectionId', 'pp_mtc_facility_id']).copy()
print(unique_oakland_match_gdf.shape)
```

```python
# status on matching and export
oakland_gdf = gpd.read_file(os.path.join(geojson_folder, 'oakland_bike_network_epsg4326.geojson'))

print('Sharedstreets matched {} out of {} oakland links'.format(
    unique_oakland_match_gdf.pp_mtc_facility_id.nunique(),
    oakland_gdf.mtc_facility_id.nunique()))

unique_oakland_match_gdf.drop(columns=['gisReferenceId', 'gisGeometryId', 'gisTotalSegments', 'gisSegmentIndex', 'gisFromIntersectionId', 'gisToIntersectionId',
                                       'startSideOfStreet', 'endSideOfStreet', 'sideOfStreet', 'score', 'matchType'], inplace=True)

unique_oakland_match_gdf.to_file(os.path.join(shst_clean_folder, 'oakland_bike_network_epsg4326_matched.geojson'),
                                 driver = "GeoJSON")
```
